```python
#all important imports

from PyPDF2 import PdfReader
from langchain.text_splitter import CharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from dotenv import load_dotenv
from langchain.vectorstores import FAISS
import os
from langchain.chains.question_answering import load_qa_chain
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
pdf='/content/test.pdf'. #store pdf file in your root in colab
```

```python
#main tool that reads the file

pdf_reader = PdfReader(pdf)
```

```python
#extracting the text and converting into chunks

text = ""
for page in pdf_reader.pages:
  text += page.extract_text()

            # Split into chunks
  text_splitter = CharacterTextSplitter(
                separator="\n",
                chunk_size=1000,
                chunk_overlap=200,
                length_function=len
            )
  chunks = text_splitter.split_text(text)
```

```python
chunks
```

```python
# using hugging face model to create embedding //use any other nice models (google, openai)

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")

#define your own vector store
KNOWLEDGE_BASE = FAISS.from_texts(chunks, embeddings)
```

```python
#define the llm , can use openai, grok, or anyother

GOOGLE_API_KEY=""

llm = ChatGoogleGenerativeAI(
            model="gemini-2.0-flash-lite",
            google_api_key=GOOGLE_API_KEY,
            temperature=0,
            max_retries=2,
        )



```

```python
#search the most relevant from pdf and load into llm to get the reponse

user_question =input("Enter your question")

docs = KNOWLEDGE_BASE.similarity_search(user_question)
chain = load_qa_chain(llm, chain_type="stuff")

response = chain.run(input_documents=docs, question=user_question)
```
